```python
import pickle
import matplotlib.pyplot as plt
```

```python
files_to_load = ['rewards.pkl', 'portfolio.pkl', 'training_loss.pkl', 'eval_port.pkl']
```

```python
data_dict = {}
for file in files_to_load:
    try:
        with open(file, 'rb') as f:
            data = pickle.load(f)
        data_dict[file] = data
    except:
        print('Failed to read ', file)
```

```python
plt.plot(data_dict['eval_port.pkl'])
```
